```python
import plotly.graph_objects as go
import pandas as pd
from matplotlib.colors import to_rgba

def sankey_condition_family(df, title="Condition–TF Family Sankey"):
    """
    Sankey plot showing the number of TFs from each family per condition.
    """

    # --- prepare data ---
    df = df.reset_index()[['OLN','Family','Groups_corrected']].drop_duplicates()
    df['Groups_corrected'] = df['Groups_corrected'].str.replace('_pval', '', regex=False)

    # split multiple groups and explode
    df['Groups_corrected'] = df['Groups_corrected'].str.split('&')
    df_exploded = df.explode('Groups_corrected')
    df_exploded['Groups_corrected'] = df_exploded['Groups_corrected'].str.strip()

    # count TFs per condition-family pair
    counts = df_exploded.groupby(['Groups_corrected','Family']).size().reset_index(name="count")

    # --- get sorted unique conditions and families ---
    total_tf_per_condition = counts.groupby("Groups_corrected")["count"].sum().sort_values(ascending=False)
    sorted_conditions = total_tf_per_condition.index.tolist()
    total_tf_per_family = counts.groupby("Family")["count"].sum().sort_values(ascending=False)
    sorted_families = total_tf_per_family.index.tolist()

    # --- node list ---
    cond_labels = [f"{c} ({total_tf_per_condition[c]})" for c in sorted_conditions]
    fam_labels  = [f"{f} ({total_tf_per_family.get(f, 0)})" for f in sorted_families]
    all_nodes = cond_labels + fam_labels
    node_idx = {name: i for i, name in enumerate(all_nodes)}

    # --- build links ---
    links = []
    for _, row in counts.iterrows():
        cond = row["Groups_corrected"]
        fam = row["Family"]
        value = row["count"]

        col = extended_family_color_dict.get(fam, (0.5,0.5,0.5,0.6))  # default grey
        rgba = to_rgba(col)  # ensures proper rgba tuple
        color_str = f"rgba({int(rgba[0]*255)},{int(rgba[1]*255)},{int(rgba[2]*255)},{rgba[3]:.2f})"

        links.append({
            "source": node_idx[f"{cond} ({total_tf_per_condition[cond]})"],
            "target": node_idx[f"{fam} ({total_tf_per_family.get(fam, 0)})"],
            "value": value,
            "color": color_str
        })

     # --- assign fixed positions ---
    n_cond = len(sorted_conditions)
    n_fam = len(sorted_families)
    x_pos, y_pos = [], []

        # left column (conditions)
    for i, _ in enumerate(sorted_conditions):
        x_pos.append(0.001)
        if n_cond > 1:
            y_pos.append((i-0.005) / (n_cond - 1))# evenly spaced top→bottom
        else:
            y_pos.append(0.5)

    # right column (pathways)
    for i, _ in enumerate(sorted_families):
        x_pos.append(0.999)
        if n_fam > 1:
            y_pos.append((i-0.001) / (n_fam - 1))
        else:
            y_pos.append(0.5)
    
    # --- unpack ---
    sources = [l["source"] for l in links]
    targets = [l["target"] for l in links]
    values = [l["value"] for l in links]
    colors = [l["color"] for l in links]

    # --- build figure ---
    fig = go.Figure(data=[go.Sankey(
        arrangement="fixed",
        node=dict(
            pad=60,
            thickness=14,
            line=dict(color="black", width=0.5),
            label=all_nodes,
            color=["rgba(160,160,160,0.9)"]*len(all_nodes),
            x=x_pos,
            y=y_pos
        ),
        link=dict(
            source=sources,
            target=targets,
            value=values,
            color=colors,
            hovertemplate="Condition → TF Family<br>%{source.label} → %{target.label}<br>"
                          "Count: %{value}<extra></extra>"
        )
    )])
    
    fig.update_layout(
        title=title,
        font_size=12,
        margin=dict(l=10, r=10, t=40, b=10),
        width=1200,
        height=1000
    )

    return fig
```
